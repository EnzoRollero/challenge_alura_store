### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda2.sample(5)
```

#1. Análisis de facturación



```python
#sumando el precio de las ventas e imprimiendo
facturacion_tienda = tienda['Precio'].sum()
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()


print(f'Facturación Tienda 1: ${facturacion_tienda:.2f}')
print(f'Facturación Tienda 2: ${facturacion_tienda2:.2f}')
print(f'Facturación Tienda 3: ${facturacion_tienda3:.2f}')
print(f'Facturación Tienda 4: ${facturacion_tienda4:.2f}')

#facturacion total
facturacion_total = facturacion_tienda + facturacion_tienda2 + facturacion_tienda3 + facturacion_tienda4
print(f'Facturación Total: ${facturacion_total:.2f}')
```

# 2. Ventas por categoría

```python
#FUNCION PARA CALCULAR CANTIDAD DE PRODUCTOS VENDIDOS POR CATEGORIA
def ventas_categoria (tienda_cat, numero):
  ventas_cat = tienda_cat['Categoría del Producto'].value_counts()
  print(f'\nVentas por categoría en la tienda {numero}')
  print(ventas_cat)

#funcion para calcular el monto total de los productos vendidos por categoria
def monto_categoria (tienda_mon, numero):
  monto_cat = tienda_mon.groupby('Categoría del Producto')['Precio'].sum()
  print(f'\nMonto total de ventas por categoria en la tienda {numero}')
  print(monto_cat)

#mostrando resultados para cada tienda
ventas_categoria(tienda, 1)
monto_categoria(tienda, 1)

ventas_categoria(tienda2, 2)
monto_categoria(tienda2, 2)

ventas_categoria(tienda3, 3)
monto_categoria(tienda3, 3)

ventas_categoria(tienda4, 4)
monto_categoria(tienda4, 4)
```

# 3. Calificación promedio de la tienda


```python
#Funcion para calcular el promedio de la calificacion de una tienda
def calificacion_tienda (tienda_cal, numero):
  calificacion_tienda = round(tienda_cal['Calificación'].mean(), 2)
  print(f'\nCalidicación promedio de la tienda {numero}: {calificacion_tienda}')

calificacion_tienda(tienda, 1)
calificacion_tienda(tienda2, 2)
calificacion_tienda(tienda3, 3)
calificacion_tienda(tienda4, 4)
```

# 4. Productos más y menos vendidos

```python
def ventas_tienda (tienda_ini, numero):
  #Obtengo el ingreso total por tienda y producto, luego ordeno en forma de mayor a menor
    ventas = tienda_ini.groupby('Producto')['Precio'].sum().reset_index(name='Ingresos Totales')
    ventas_orden = ventas.sort_values(by='Ingresos Totales', ascending=False)
  #Imprimo los 3 productos que mas ingresos generaron y lo hago con metodo to_string() para formato de salida mas limpio
    print(f'\n Tienda {numero} - TOP 3 Productos que MÁS ingresos generaron:')
    print(ventas_orden.head(3).to_string(index=False, columns=['Producto', 'Ingresos Totales'], formatters={'Ingresos Totales': '${:,.2f}'.format} ))
  #Imprimo los 3 productos que menos ingresos generaron y lo hago con metodo to_string() para formato de salida mas limpio
    print(f'\n Tienda {numero} - TOP 3 Productos que MENOS ingresos generaron:')
    print(ventas_orden.tail(3).to_string(index=False, columns=['Producto', 'Ingresos Totales'], formatters={'Ingresos Totales': '${:,.2f}'.format}))

#Llamo a la funcion con cada una de las tiendas
ventas_tienda(tienda, 1)
ventas_tienda(tienda2, 2)
ventas_tienda(tienda3, 3)
ventas_tienda(tienda4, 4)
```

# 5. Envío promedio por tienda

```python


def envio_tienda (tienda_env, numero):
  #Aseguro que el costo del envio es numerico
  tienda_env['Costo de envío'] = tienda_env['Costo de envío'].astype(float)
  #Calculo el costo promedio de envío de cada tienda
  envio_promedio = tienda_env['Costo de envío'].mean()

  #muestro el resultado
  print(f'\nEl costo promedio de envío de la Tienda {numero} es: ${envio_promedio:.2f}')


envio_tienda(tienda, 1)
envio_tienda(tienda2, 2)
envio_tienda(tienda3, 3)
envio_tienda(tienda4, 4)

```

```python
# 1. Aseguramos que la columna 'Fecha de Compra' sea datetime
tienda_total['Fecha de Compra'] = pd.to_datetime(tienda_total['Fecha de Compra'], dayfirst=True)

#Si consideramos que el costo de envio lo cubre la tienda debo restarselo al precio del producto
tienda_total['Ganancia'] = tienda_total['Precio'] - tienda_total['Costo de envío']

#Agrupo por mes y por tienda, para sumar la ganancia
ganancia_mensual = tienda_total.groupby([tienda_total['Fecha de Compra'].dt.to_period('M'),#agrupacion por mes
                                         'Tienda_ID'])['Ganancia'].sum().reset_index()

#Se convierte la fecha de la ganancia en str para poder graficarla
ganancia_mensual['Fecha de Compra'] = ganancia_mensual['Fecha de Compra'].astype(str)

#grafico la serie temporal
plt.figure(figsize=(12,6))
#Usamos 'hue=Tienda_ID' para que cada tienda genere una linea diferente
sns.lineplot(data=ganancia_mensual, x='Fecha de Compra', y='Ganancia', hue='Tienda_ID', marker='o')
plt.title('Ganancia Mensual por Tienda', fontsize=15)
plt.ylabel('Ganancia Total Mensual (USD)')
plt.xlabel('Mes')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 1. Aseguramos que la columna 'Fecha de Compra' sea datetime
tienda_total['Fecha de Compra'] = pd.to_datetime(tienda_total['Fecha de Compra'], dayfirst=True)

#Calculo ventas mensuales por cada tienda
#tengo que agrupar por mes y por tienda y luego sumar el precio
ventas_mensuales = tienda_total.groupby([tienda_total['Fecha de Compra'].dt.to_period('M'), 'Tienda_ID'])['Precio'].sum().reset_index()

#Renombro las columnas de ventas_mensuales para mejor comprension y para que coincida el nombre de la columna 'Tienda'
ventas_mensuales.columns = ['Fecha', 'Tienda', 'Ingresos Mensuales']

#Convertir el indice de periodo a timestamp para trazar
ventas_mensuales['Fecha'] = ventas_mensuales['Fecha'].dt.to_timestamp()
ventas_mensuales = ventas_mensuales.set_index('Fecha')

# 2. Defino la paleta de colores para destacar la Tienda 4
# Mapeo los colores a las tiendas: Tienda 4 debe ser roja
color_map = {'Tienda 1': '#1f77b4', 'Tienda 2': '#ff7f0e', 'Tienda 3': '#2ca02c', 'Tienda 4': 'red'}

plt.figure(figsize=(12,6))

for tienda in ventas_mensuales['Tienda'].unique():
  data_tienda = ventas_mensuales[ventas_mensuales['Tienda'] == tienda].copy()
  data_tienda['Media_Movil_6M'] = data_tienda['Ingresos Mensuales'].rolling(window=6).mean()

  #grafico usando el color correspondiente para la tienda
  sns.lineplot(data=data_tienda, x=data_tienda.index, y='Media_Movil_6M',
               label=tienda, marker='', color=color_map[tienda])

plt.title('Tendencia Suavizada de Ventas Mensuales (Media Móvil 6M)', fontsize=15)
plt.xlabel('Fecha')
plt.ylabel('Ingreso Promedio Suavizado (USD)')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()
```
